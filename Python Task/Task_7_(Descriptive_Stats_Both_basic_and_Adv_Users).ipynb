- If the number is even, find the two middle terms using the formula n/2 and (n/2) + 1.

**Measure of Dispersion**

Dispersion in statistics is a way of describing how to spread out a set of data is.

The measures of dispersion contain almost the same unit as the quality being measured. There are many Measures of dispersion found that helps us to get more insights into the data.
1. Range
2. Variance
3. Standard Deviation 
4. Skewness
5. IQR

image.png

**Types of Measure of Dispersion**

1. **Absolute Measure Dispersion**
- **Range** : Given a data set, the range can be defined as the difference between the maximum value and the minimum value.

- **Variance** : The average squared deviation from the mean of the given data set is known as the variance. This measure of dispersion checks the spread of the data about the mean.

image.png

- **Standard Deviation** : The square root of the variance gives the standard deviation. Thus, the standard deviation also measures the variation of the data about the mean.

- **Mean Deviation** : The mean deviation gives the average of the data's absolute deviation about the central points. These central points could be the mean, median, or mode.

- **Quartile Deviation** : Quartile deviation can be defined as half of the difference between the third quartile and the first quartile in a given data set.

2. **Relative Measure Dispersion** 

- **Coefficient of Range**: It is the ratio of the difference between the highest and lowest value in a data set to the sum of the highest and lowest value.

- **Coefficient of Variation**: It is the ratio of the standard deviation to the mean of the data set. It is expressed in the form of a percentage.

- **Coefficient of Mean Deviation**: This can be defined as the ratio of the mean deviation to the value of the central point from which it is calculated.

- **Coefficient of Quartile Deviation**: It is the ratio of the difference between the third quartile and the first quartile to the sum of the third and first quartiles.

**Pandas implementation of measure of dispersion**

```python
# variance

df['Emi_or_Rent_Amt'].var()
```

```python
n = len(df['Emi_or_Rent_Amt'])
mean = sum(df['Emi_or_Rent_Amt'])/n
deviations = [(x-mean)**2 for x in df['Emi_or_Rent_Amt']]
var = sum(deviations)/(n-1)
print(var)
```

```python
# standard Deviation = Standard deviation is square root of variance


n = len(df['Emi_or_Rent_Amt'])
mean = sum(df['Emi_or_Rent_Amt'])/n
deviations = [(x-mean)**2 for x in df['Emi_or_Rent_Amt']]
var = (sum(deviations)/n)**0.5
print(var)
```

**Correlation**

Correlation explains how one or more variables are related to each other. These variables can be input data features which have been used to forecast our target variable.

Correlation, statistical technique which determines how one variables moves/changes in relation with the other variable. It gives us the idea about the degree of the relationship of the two variables. It’s a bi-variate analysis measure which describes the association between different variables. In most of the business it’s useful to express one subject in terms of its relationship with others.


There are three correlation:

**Positive Correlation** : Two features (variables) can be positively correlated with each other. It means that when the value of one variable increase then the value of the other variable(s) also increases.

**Negative Correlation** : Two features (variables) can be negatively correlated with each other. It means that when the value of one variable increase then the value of the other variable(s) decreases.

**No Correlation** : Two features (variables) are not correlated with each other. It means that when the value of one variable increase or decrease then the value of the other variable(s) doesn’t increase or decreases.

**Normal Distribution**

The Normal Distribution is also known as Gassian and Gauss distribution, is a continuous probability distribution. It has a bell-shaped Curve that is symmetrical from the mean point to both halves of the curve.

image.png

image.png

**Features of Normal Distribution**

- In a Normal Distribution It is Symmetric about the mean--half the values fall below the mean and half above the mean.
- The mean, median and mode are equal.
- The distribution can be described by two values: the mean and standard deviation

**Positive Skewed and Negatively Skewed Normal Distribution**

In simple words, Skewness is the measure of how much the probability distribution of a random variable deviates from the normal distribution. 
- Positive Skewed distribution is the distribution with the tail on its right side. The value of skewness for a positively skewed distribution is greater than zero.

**Mode** < **Median** < **Mean**

image.png

- **Negative Skewed** distribution is the distribution with the tail on its left side. The value of skewness for a negatively skewed distribution is less than zero.

**Mode > Median > Mean**
image.png

**Effect Of Mean, Median, Mode on Skewness**

Generally if the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode (Mean<Median<Mode). If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean(Mode>Median>Mean).

**Explain QQ Plot and show the implementation of the same**


Q-Q plots are also known as Quantile-Quantile plots. As the name suggests, they plot the quantiles of a sample distribution against quantiles of a theoretical distribution. Doing this helps us determine if a dataset follows any particular type of probability distribution like normal, uniform, exponential.

```python
from scipy import stats

stats.probplot(df['Annual_HH_Income'], dist='norm', plot = plt)
plt.show()
```

```python
# for all the columns plotting Q-Q plot
columns = ['Mthly_HH_Income', 'Mthly_HH_Expense', 'No_of_Fly_Members',
       'Emi_or_Rent_Amt', 'Annual_HH_Income',
       'No_of_Earning_Members']
for col in columns:
  stats.probplot(df[col], dist='norm', plot = plt)
  plt.show()
```

Here In these plotting some of the plots are uniformly distributed and some of the Exponential Distribution.

**Explain Box Cox and show the implementation of the same**

The Box-Cox transformation transforms our data so that it closely resembles a normal distribution.   

A Box Cox transformation is a transformation of non-normal dependent variables into a normal shape. Normality is an important assumption for many statistical techniques; if your data isn’t normal, applying a Box-Cox means that you are able to run a broader number of tests.

```python
sns.distplot(df['Annual_HH_Income'], hist=False, kde=True, kde_kws={'shade':True, 'linewidth':2}, label= 'Non-Normal', color='green')
```

```python
# fitted_data, fitted_lambda = stats.boxcox(original_data)

fitted_data, fitted_lambda=stats.boxcox(df['Annual_HH_Income'])

fitted_data
```

```python
print(f"Lambda value used for Transformation: {fitted_lambda}")
```

```python
sns.distplot(fitted_data, hist=False, kde=True, kde_kws={'shade':True, 'linewidth':2}, label='Non-Normal', color='green')
```

Here we can the non-normal distribution was converted into normal distribution or rather close to normal using the scipy.stats.boxcox().
