```python
# finding outlier of using IQR method

def outlier_detect_using_iqr(data, col):
  q1 = data[col].quantile(0.25)
  q3 = data[col].quantile(0.75)
  interquartile_range = q3 - q1

  lower_bound = q1 - (1.5*interquartile_range)
  upper_bound = q3 - (1.5*interquartile_range)

  outliers = []

  for val in data[col]:
    if val < lower_bound or val > upper_bound:
      outliers.append(val)
  
  return outliers,len(outliers)
```

```python
outlier_detect_using_iqr(data, 'Salary')
```

```python
# outliers for 12percentage
sns.boxplot(data = outlier_detect_using_iqr(data, '12percentage'))
plt.title("Outliers on 12percentage")

```

```python
#Outliers are affecting "Specialization" data

def outlier_detect_using_iqr(data, col):
  q1 = data[col].quantile(0.25)
  q3 = data[col].quantile(0.75)
  interquartile_range = q3 - q1

  lower_bound = q1 - (1.5*interquartile_range)
  upper_bound = q3 + (1.5*interquartile_range)

  outliers = []

  for val in data[col]:
    if val < lower_bound or val > upper_bound:
      outliers.append(val)
  
  return len(outliers)
```

```python
print(outlier_detect_using_iqr(data, 'extraversion'))
```

```python
for column in ['extraversion','openess_to_experience','Domain','agreeableness','conscientiousness','10percentage' ,'Quant','12percentage','English', 'ComputerProgramming', 'Logical']:
  print(outlier_detect_using_iqr(data, column))
```
